### **Trocar o nome da coluna 'droga' para 'composto' com mapa e rename:**

```python
mapa = {'droga': 'composto'}
dados.rename(columns=mapa, inplace=True)
dados.head()
```

### **Identificando os 5 'compostos' mais frequentes:**

```python
cod_compostos = dados['composto'].value_counts()[:5] #este cria uma nova dataframe, n√£o pode ser usada no query
print(cod_compostos)
cod_compostos = cod_compostos.index #esta cria uma lista de √≠ndices e pode ser usada no query
cod_compostos
```

```python
dados.query('composto in @cod_compostos')
```

### **Criando um gr√°fico do tipo countplot para os 5 'compostos' mais frequentes:**

```python
# importando o Seaborn:
import seaborn as sns
sns.set()
```

```python
plt.figure(figsize=(10,8))
ax = sns.countplot(x='composto', data=dados.query('composto in @cod_compostos')).set_title('Top 5 Compostos')
plt.show()
```

###Desafio 01: Ordenar o gr√°fico countplot

```python
plt.figure(figsize=(10,8))
sns.countplot(x='composto', data=dados.query('composto in @cod_compostos'), order=cod_compostos).set_title('Top 5 Compostos')
plt.show()
```

###Desafio 02: Melhorar a visualiza√ß√£o alterando tamanho da fonte...

```python
plt.figure(figsize=(20,8))
sns.countplot(x='composto', data=dados.query('composto in @cod_compostos'), order=cod_compostos)
plt.title('Top 5 Compostos', fontdict={'fontsize': 30, 'fontfamily': 'serif', 'fontweight': 'bold'})
plt.ylabel('Frequ√™ncias', fontdict={'fontsize': 20})
plt.xlabel('Compostos mais Frequentes', fontdict={'fontsize': 20})
plt.show()
```

### **Identificando informa√ß√µes (m√°ximo, m√≠nimo, m√©dia) com rela√ß√£o √†s express√µes g√™nicas ('g') e criando histogramas:**
