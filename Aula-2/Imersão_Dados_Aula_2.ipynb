### **Base de dados atualizada:**

### **Trocar o nome da coluna 'droga' para 'composto' com mapa e rename:**

```python
mapa = {'droga': 'composto'}
dados.rename(columns=mapa, inplace=True)
dados.head()
```

### **Identificando os 5 'compostos' mais frequentes:**

```python
cod_compostos = dados['composto'].value_counts()[:5] #este cria uma nova dataframe, não pode ser usada no query
print(cod_compostos)
cod_compostos = cod_compostos.index #esta cria uma lista de índices e pode ser usada no query
cod_compostos
```

```python
dados.query('composto in @cod_compostos')
```

### **Criando um gráfico do tipo countplot para os 5 'compostos' mais frequentes:**

```python
# importando o Seaborn:
import seaborn as sns
sns.set()
```

```python
plt.figure(figsize=(10,8))
ax = sns.countplot(x='composto', data=dados.query('composto in @cod_compostos')).set_title('Top 5 Compostos')
plt.show()
```

###Desafio 01: Ordenar o gráfico countplot

```python
plt.figure(figsize=(10,8))
sns.countplot(x='composto', data=dados.query('composto in @cod_compostos'), order=cod_compostos).set_title('Top 5 Compostos')
plt.show()
```

###Desafio 02: Melhorar a visualização alterando tamanho da fonte...

```python
plt.figure(figsize=(20,8))
sns.countplot(x='composto', data=dados.query('composto in @cod_compostos'), order=cod_compostos)
plt.title('Top 5 Compostos', fontdict={'fontsize': 30, 'fontfamily': 'serif', 'fontweight': 'bold'})
plt.ylabel('Frequências', fontdict={'fontsize': 20})
plt.xlabel('Compostos mais Frequentes', fontdict={'fontsize': 20})
plt.show()
```

### **Identificando informações (máximo, mínimo, média) com relação às expressões gênicas ('g') e criando histogramas:**

```python
print('Quantidade de expressões gênicas g0 diferentes:')
print(len(dados["g0"].unique()))
print('Valor mínimo de g0:')
print(dados['g0'].min())
print('Valor médio de g0:')
print(dados['g0'].mean())
print('Valor máximo de g0:')
print(dados['g0'].max())
```

```python
dados['g0'].hist(bins=100).set_title('Histograma de frequências dos valores de g0')
plt.show()
```

```python
dados['g19'].hist(bins=100).set_title('Histograma de frequências dos valores de g19')
plt.show()
```

```python
dados.loc[:, 'g0':'g771'].describe()
```

```python
descricao_dados_g = dados.loc[:, 'g0':'g771'].describe().T
descricao_dados_g.head()
```

```python
print('Os valores míminos dos mínimos é:')
print(descricao_dados_g['min'].min())
print('O valor médio das médias é:')
print(descricao_dados_g['mean'].mean())
print('O valor máximo dos máximos é:')
print(descricao_dados_g['max'].max())
```
